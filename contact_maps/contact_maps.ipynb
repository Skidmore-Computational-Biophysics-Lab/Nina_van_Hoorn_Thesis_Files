```python
%matplotlib inline
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt
from contact_map import ContactFrequency, ContactDifference
from contact_map import AtomMismatchedContactDifference
```

```python
minA = 1
maxA = 130
minB = 191
maxB = 290
traj_ctd_list = []
```

# Last 300 Seconds VIF-A3F_GLU_mut

```python
traj_list = ["VCBC-A3F_GLU_mut_sims1-8_last300.mdcrd"]

traj_new = []
for suffix in traj_list:
    traj_new.append(md.load_netcdf("/data/nvanhoorn/A3F_newstructure/VCBC_A3F_GLU_mut/mdcrd/" + suffix, 
                          top="/data/nvanhoorn/A3F_newstructure/VCBC_A3F_GLU_mut/VCBC-A3F_GLU_mut_nowat.prmtop", stride=10))
```

```python
traj_new[0]
```

```python
topology = traj_new[0].topology
topology
```

```python
traj_combined = []
for file in traj_new:
    traj_combined = md.join(file, traj_combined)
```

```python
GLU_mut_combined = ContactFrequency(traj_combined) #maybe specify residues
```

```python
%%time
fig, ax = GLU_mut_combined.residue_contacts.plot(nrows=1, ncols=1, figsize=(8, 8))

my_xticks = []
current_xticks = []

for x in range(int(minA), int(maxA)+1, 3):
    current_xticks.append(x+0.5)
    my_xticks.append(x)
    
my_yticks = []
current_yticks = []
for y in range(int(minB), int(maxB)+1, 3):
    current_yticks.append(y+0.5)
    my_yticks.append(y)

new_Xticks_L = my_xticks[::2]
new_Yticks_L = my_yticks[::2]
new_Xticks = current_xticks[::2]
new_Yticks = current_yticks[::2]

plt.xticks(new_Xticks, new_Xticks_L, fontsize = 10, rotation=-50)
plt.yticks(new_Yticks, new_Yticks_L, fontsize = 10, rotation=45)

ax.set_xlim(minA,maxA)
ax.set_ylim(minB,maxB)
plt.xlabel("A3F Residues", fontsize=18)
plt.ylabel("Vif Residues", fontsize=18)
plt.title('Mutated HIV-VIF A3F Contact Map', fontsize=18)

fig.savefig('/data/nvanhoorn/A3F_newstructure/VCBC_A3F_GLU_mut/analysis/contact_map/updated_contact_maps/non-reordered/mut_contacts.png', bbox_inches = "tight", dpi= 2000)

plt.show()
```
