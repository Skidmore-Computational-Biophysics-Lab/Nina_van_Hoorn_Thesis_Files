```python
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker
```

## PCA for GLU mutations

```python
PCA1 = pd.read_table ('GLU_mut_project_onto_evecs.dat', sep = '\\s+') 
PCA2 = pd.read_table ('VCBC_A3F_project_onto_evecs.dat', sep = '\\s+') 
pdb_PCA = pd.read_table ('pdb_project_onto_evecs.dat', sep = '\\s+')
```

```python
print(max(PCA1['Mode1'].max(), PCA2['Mode1'].max()))
print(min(PCA1['Mode1'].min(), PCA2['Mode1'].min()))
print("---------------")

print(max(PCA1['Mode2'].max(), PCA2['Mode2'].max()))
print(min(PCA1['Mode2'].min(), PCA2['Mode2'].min()))
print("---------------")

print(max(PCA1['Mode3'].max(), PCA2['Mode3'].max()))
print(min(PCA1['Mode3'].min(), PCA2['Mode3'].min()))
print("---------------")

print(max(PCA1['Mode4'].max(), PCA2['Mode4'].max()))
print(min(PCA1['Mode4'].min(), PCA2['Mode4'].min()))
```

```python
print((PCA1['Mode1'].max(), PCA2['Mode1'].max()))
print((PCA1['Mode1'].min(), PCA2['Mode1'].min()))
print("---------------")

print((PCA1['Mode2'].max(), PCA2['Mode2'].max()))
print((PCA1['Mode2'].min(), PCA2['Mode2'].min()))
print("---------------")

print((PCA1['Mode3'].max(), PCA2['Mode3'].max()))
print((PCA1['Mode3'].min(), PCA2['Mode3'].min()))
print("---------------")

print((PCA1['Mode4'].max(), PCA2['Mode4'].max()))
print((PCA1['Mode4'].min(), PCA2['Mode4'].min()))
```

```python
binedges = np.arange(-500,230,1) # create array for bin edges
VCBC_V137Gly = np.histogram2d(PCA1['Mode1'], PCA1['Mode2'], 
                         bins=binedges)
max_shade = VCBC_V137Gly[0].max()
PC1_2_colormap = plt.figure(figsize=(14.5,9))
plt.subplot(1, 2, 1)
PC2_VCBC_V137Gly, PC1_VCBC_V137Gly = np.meshgrid(binedges,binedges) 
plt.pcolormesh(PC1_VCBC_V137Gly,PC2_VCBC_V137Gly,VCBC_V137Gly[0], cmap = 'Greys', vmax = max_shade)
plt.axis('image')
plt.xlim([-310,240])
plt.ylim([-485,160])
def fmt(x, pos):
    a, b = '{:.2e}'.format(x).split('e')
    b = int(b)
    return r'${} \times 10^{{{}}}$'.format(a, b)
plt.colorbar(shrink = 0.50, format=ticker.FuncFormatter(fmt))
plt.scatter(x = pdb_PCA["Mode1"][0], y = pdb_PCA["Mode2"][0], color='k', 
         marker='x', s = 150, linewidth = 3, label = "VCBC pdb")
plt.title('GLU_mutation', size= 18) 
plt.xlabel('PC1',size= 16) 
plt.ylabel('PC2',size= 16)
```

```python
binedges = np.arange(-275,165,1) # create array for bin edges
VCBC_V137Gly = np.histogram2d(PCA1['Mode3'], PCA1['Mode4'], 
                         bins=binedges)
max_shade = VCBC_V137Gly[0].max()
PC1_2_colormap = plt.figure(figsize=(14.5,9))
plt.subplot(1, 2, 1)
PC2_VCBC_V137Gly, PC1_VCBC_V137Gly = np.meshgrid(binedges,binedges) 
plt.pcolormesh(PC1_VCBC_V137Gly,PC2_VCBC_V137Gly,VCBC_V137Gly[0], cmap = 'Greys', vmax = max_shade)
plt.axis('image')
plt.xlim([-125,90])
plt.ylim([-120,130])
def fmt(x, pos):
    a, b = '{:.2e}'.format(x).split('e')
    b = int(b)
    return r'${} \times 10^{{{}}}$'.format(a, b)
plt.colorbar(shrink = 0.50, format=ticker.FuncFormatter(fmt))
plt.scatter(x = pdb_PCA["Mode3"][0], y = pdb_PCA["Mode4"][0], color='k', 
         marker='x', s = 150, linewidth = 3, label = "VCBC pdb")
plt.title('GLU_mutation', size= 18) 
plt.xlabel('PC3',size= 16) 
plt.ylabel('PC4',size= 16)
```
